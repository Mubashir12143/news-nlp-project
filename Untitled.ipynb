```python
import pandas as pd
import feedparser
```

```python
sources = {
    "BBC": "http://feeds.bbci.co.uk/news/rss.xml",
    "Reuters": "https://feeds.reuters.com/reuters/worldNews",
    "AlJazeera": "https://www.aljazeera.com/xml/rss/all.xml"
}
```

```python
def fetch_news():
    all_news = []

    for source, url in sources.items():
        feed = feedparser.parse(url)

        for entry in feed.entries:
            headline = entry.title

            all_news.append({
                "headline": headline,
                "source": source
            })

    return pd.DataFrame(all_news)
```

```python
df = fetch_news()
df.head()
```

```python
print("Total news:", len(df))
```

```python
df.head()
```

```python
df = df.drop_duplicates(subset="headline")
```

```python
df = df[df["headline"].str.len() > 20]
```

```python
df = df.reset_index(drop=True)
```

```python
print("Clean dataset size:", len(df))
df.head()
```

```python
df["headline"] = df["headline"].str.replace(r"[^a-zA-Z0-9 ]", "", regex=True)
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

analyzer = SentimentIntensityAnalyzer()
```

```python
def get_sentiment(text):
    score = analyzer.polarity_scores(text)["compound"]
    
    if score >= 0.05:
        return "Positive"
    elif score <= -0.05:
        return "Negative"
    else:
        return "Neutral"
```

```python
df["sentiment"] = df["headline"].apply(get_sentiment)
df.head()
```

```python
df["sentiment"].value_counts()
```

```python
df["score"] = df["headline"].apply(lambda x: analyzer.polarity_scores(x)["compound"])
df.head()
```
